```python
import sys
import requests
import zipfile
import urllib3
import numpy as np

from pathlib import Path
from os import getcwd, makedirs
from os.path import join, basename, isdir

sys.path.append(str(Path(getcwd()).parent))
current_dir = getcwd()
print(current_dir)

import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 200)

urllib3.disable_warnings()
```

```python
class Dados():
    def leituraDados(entrada):
        caminho_entrada = join(Path(getcwd()).parent,"dados","microdados_Enade_2019_LGPD", "2. DADOS")
        DataFrame = pd.read_csv(join(caminho_entrada,str(entrada).split(",")[0]), sep = ';', low_memory = False).reset_index()
        count = 0
        for arquivo in str(entrada).replace(" ","").split(","):
            if count == 0:
                count = count+1
            else:
                self = DataFrame.merge(pd.read_csv(join(caminho_entrada,arquivo), sep = ';', low_memory = False).reset_index(), how = 'left', on = 'index')
        return DataFrame.set_index('index')

    def agrupaDados(dados):
        return dados.groupby([dados.columns[1],dados.columns[2]])[dados.columns[0]].count().sort_values(ascending=False).unstack()

    def baixarDados(url, endereco=None):
        if not isdir(join(Path(getcwd()).parent,"dados")):
            makedirs(join(Path(getcwd()).parent,"dados"))
        if endereco is None:
            endereco = join(Path(getcwd()).parent,"dados", basename(url.split("/")[4]))
        resposta = requests.get(url, stream=True, verify=False)
        if resposta.status_code == requests.codes.OK:
            with open(endereco, 'wb') as novo_arquivo:
                    for parte in resposta.iter_content(chunk_size=256):
                        novo_arquivo.write(parte)
            print("Download finalizado. Arquivo salvo em: {}".format(endereco))
            with zipfile.ZipFile(join(Path(getcwd()).parent,"dados", basename(url.split("/")[4])), 'r') as zip_ref:
                zip_ref.extractall(join(Path(getcwd()).parent,"dados"))
        else:
            resposta.raise_for_status()
```

```python
Dados.baixarDados("https://download.inep.gov.br/microdados/microdados_enade_2019_LGPD.zip")
dados = Dados.leituraDados("microdados2019_arq14.txt")
Dados.agrupaDados(dados)
```
